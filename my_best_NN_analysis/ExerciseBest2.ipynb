```python
import tensorflow as tf
import numpy as np
import random as rn
import tensorflow_datasets as tfds

weight_matrix_1 = np.zeros((10, 784))
bias_1 = np.zeros(10)

k = 0
l = 0
count = 0
for k in range(785):
  for l in range(10):
        count+=1
        if (k==784 and l<10):
          bias_1[l] = float(NN_data2[l+10*k])
          l += 1
        if (k<784):
          weight_matrix_1[l][k] = float(NN_data2[l+10*k])
          l += 1
          if (l==10):
            l=0
            k += 1

#print(bias_1)
```

```python
model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(input_shape=(28, 28)),
        tf.keras.layers.Dense(10, activation='relu'),
        ])

for i in range(10):
  for j in range(784):
    indices = [[j,i]]
    model.weights[0].assign(tf.tensor_scatter_nd_update(model.weights[0],indices,[weight_matrix_1[i][j]]))

for i in range(10):
  indices = [[i]]
  model.weights[1].assign(tf.tensor_scatter_nd_update(model.weights[1],indices,[bias_1[i]]))
```

```python
(ds_train, ds_test), ds_info = tfds.load(
    'mnist',
    split=['train', 'test'],
    shuffle_files=False,
    as_supervised=True,
    with_info=True,
)

def normalize_img(image, label):
  """Normalizes images: `uint8` -> `float32`."""
  return tf.cast(image, tf.float32) / 255., label

ds_train = ds_train.map(
    normalize_img, num_parallel_calls=tf.data.AUTOTUNE)
ds_train = ds_train.cache()
#ds_train = ds_train.shuffle(ds_info.splits['train'].num_examples)
ds_train = ds_train.batch(128)
ds_train = ds_train.prefetch(tf.data.AUTOTUNE)

ds_test = ds_test.map(
    normalize_img, num_parallel_calls=tf.data.AUTOTUNE)
ds_test = ds_test.batch(128)
ds_test = ds_test.cache()
ds_test = ds_test.prefetch(tf.data.AUTOTUNE)
```

```python
def index_of_max(a):
  l = a[0]
  k = 0
  for i in range(10):
    if (a[i]>l):
      l = a[i]
      k = i
  return k

def readable_print(confusion_matrix):
  for i in range(10):
    for j in range(10):
      print(int(confusion_matrix[i][j]), end="\t")
    print()

tot = 0.0

a = list(ds_train)
b = list(model.predict(ds_train,verbose=0))
confusion_matrix = np.zeros((10,10))

size = 60000
for i in range(size):
  number_of_image = int(i/128)
  number_of_image_in_batch = i%128
  x = a[number_of_image][1][number_of_image_in_batch].numpy()
  y = index_of_max(b[128*number_of_image+number_of_image_in_batch])
  if (x==y):
    tot = tot + 1.0
  confusion_matrix[x][y] = confusion_matrix[x][y]+1

den = float(size/100)
print("Train set data: ")
print("Accuracy: " + str(float(tot/den))+"%")
print("Confusion matrix: \n")
readable_print(confusion_matrix)

```
